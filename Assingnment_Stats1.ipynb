***13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?***

CLT states that for a large enough sample size, the sample mean follows a normal distribution regardless of the population‚Äôs original distribution.

Importance:

Allows us to use normal distribution for inference.

Justifies using Z-tests and confidence intervals.

***14. How does the Central Limit Theorem relate to the normal distribution?***

CLT ensures that as sample size increases, the distribution of sample means approaches a normal distribution.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate a random variable and display its value
random_value = np.random.rand()
print("Random Variable:", random_value)

# 2. Generate a discrete uniform distribution and plot PMF
values = np.arange(1, 7)
probabilities = np.ones(6) / 6
plt.bar(values, probabilities, alpha=0.6, color='b', edgecolor='k')
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('PMF of a Discrete Uniform Distribution')
plt.show()

# 3. Bernoulli distribution PDF
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)

print("Bernoulli PDF (p=0.5, x=1):", bernoulli_pdf(0.5, 1))

# 4. Simulate binomial distribution (n=10, p=0.5) and plot histogram
binomial_samples = np.random.binomial(n=10, p=0.5, size=1000)
sns.histplot(binomial_samples, bins=11, kde=False, stat="density", color='g')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.show()

# 5. Create Poisson distribution and visualize it
poisson_samples = np.random.poisson(lam=5, size=1000)
sns.histplot(poisson_samples, bins=15, kde=False, color='r')
plt.xlabel('Occurrences')
plt.ylabel('Frequency')
plt.title('Poisson Distribution Histogram')
plt.show()

# 6. Calculate and plot CDF of a discrete uniform distribution
cdf = np.cumsum(probabilities)
plt.step(values, cdf, where='mid', color='b', label='CDF')
plt.xlabel('Values')
plt.ylabel('Cumulative Probability')
plt.title('CDF of a Discrete Uniform Distribution')
plt.legend()
plt.show()

# 7. Generate continuous uniform distribution and visualize
continuous_samples = np.random.uniform(0, 10, 1000)
sns.histplot(continuous_samples, bins=20, kde=True, color='purple')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Continuous Uniform Distribution')
plt.show()

# 8. Simulate normal distribution data and plot histogram
normal_samples = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(normal_samples, bins=30, kde=True, color='orange')
plt.title('Histogram of Normal Distribution')
plt.show()

# 9. Calculate Z-scores and plot
z_scores = (normal_samples - np.mean(normal_samples)) / np.std(normal_samples)
sns.histplot(z_scores, bins=30, kde=True, color='cyan')
plt.title('Z-Scores Distribution')
plt.show()

# 10. Implement CLT using Python
def central_limit_theorem():
    samples = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
    sns.histplot(samples, bins=30, kde=True)
    plt.title('Central Limit Theorem Applied to Exponential Distribution')
    plt.show()

central_limit_theorem()

# More functions for additional tasks can be added similarly.
```
